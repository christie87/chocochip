```python
#7 
'''
Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)? 
Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent
 or Married-AF-spouse), the rest are considered bachelors.
'''
import copy
ms=copy.deepcopy(df[['marital-status','income']])
ms['marital-status']=ms['marital-status'].str.replace(r'Married.+','Married',regex=True)
ms['marital-status'] = ms['marital-status'].replace(to_replace=r"^(.(?<!Married))*?$", value='Not Married',regex=True)
ms=ms.loc[ms['income']=='>50K']
ms.groupby('marital-status')[['income']].count( )
```

```python
#8  
'''
What is the maximum number of hours a person works per week (hours-per-week feature)? 
How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?
'''
print(df['hours-per-week'].max())
print(df.loc[df['hours-per-week']==99,['hours-per-week']].count())
print((df.loc[(df['hours-per-week']==99) & (df['income']=='>50K'),['income']].count()/
       df.loc[df['hours-per-week']==99,['income']].count())*100)
```

```python
#9 
''' Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) 
for each country (native-country).What will these be for Japan?
'''
df.groupby(['income','native-country'])[['hours-per-week']].mean()
```

```python
data=df.loc[(df['native-country']=='Japan') &(df['income']=='>50K'),['hours-per-week']].mean()
data1=df.loc[(df['native-country']=='Japan') &(df['income']=='<=50K'),['hours-per-week']].mean()
print('Average hours per week over 50K in Japan is',data)
print('Average hours per week under 50K in Japan is', data1)
```
