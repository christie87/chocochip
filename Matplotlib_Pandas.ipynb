```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/sample_data/company_sales_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Exercise 1: Read Total profit of all months and show it using a line plot

x=df.loc[:,['month_number']]
y=df.loc[:,['total_profit']]
plt.plot(x, y)
plt.title('Total Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.show()
```

```python
'''Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: 
â€“Line Style dotted and Line-color should be redShow legend at the lower right location.
X label name = Month Number Y label name = Sold units number
Add a circle marker.Line marker color as readLine width should be 3'''

y1=df.loc[:,['total_units']]
plt.plot(x, y1, 'ro:',linewidth=3, label='units' )
plt.title('Total Units by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units')
plt.legend(loc='lower right')
plt.show()
```

```python
#Exercise 3: Read all product sales data and show it using a multiline plot
y2=df.iloc[:,1]
y3=df.iloc[:,2]
y4=df.iloc[:,3]
y5=df.iloc[:,4]
y6=df.iloc[:,5]
y7=df.iloc[:,6]

plt.plot(x, y2, label='facecream')
plt.plot(x, y3, label='facewash')
plt.plot(x, y4, label='toothpaste')
plt.plot(x, y5, label='bathing soap')
plt.plot(x, y6, label='shampoo')
plt.plot(x, y7, label='moisturizer')
plt.title('Product Sales by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units')
plt.legend()
plt.show()
```

```python
#Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
t=df.loc[:,'toothpaste']
plt.scatter(x,t,label='toothpaste')
plt.title(' Toothpaste Sales by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units')
plt.legend()
plt.grid()
plt.show()
```

```python
#Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

df.plot('month_number',['facecream','facewash'], kind='bar')
plt.title(' Item Sales by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units')
plt.legend()
plt.show()
```

```python
face=df['facecream']
wash=df['facewash']
m=df['month_number']
plt.bar(m,face, label='facecream',align='edge',width=0.4)
plt.bar(m,wash,label='facewash',align='edge',width=-0.4)
plt.title(' Item Sales by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units')
plt.legend()
plt.show()
```

```python
#Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges
p=df['total_profit']
plt.hist(p,bins=4)
plt.title(' Distribution of Total Profit')
plt.ylabel('Number of Months')
plt.xlabel('Total Profit Range')
plt.show()
```

```python
bins=pd.cut(df['total_profit'],4)
bins.value_counts()
```

```python
#Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
ts=df.iloc[:,1:7].sum()
plt.pie(ts,autopct='%1.1f%%',labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
plt.title(' Item Sales by Month')
plt.show()
```

```python
#Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot
soap=df.bathingsoap
fwash=df.facewash
plt.subplot(2,1, 1)
plt.plot(x, soap, 'ro-')
plt.title('Bathing Soap and Facewash Sales per Month')
plt.ylabel('Bathing Soap')

plt.subplot(2,1, 2)
plt.plot(x, fwash, 'y.-')
plt.xlabel('Months')
plt.ylabel('Facewash')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
fig.suptitle('Bathing Soap and Facewash Sales per Month')

ax1.plot(x, soap, 'ro-')
ax1.set_ylabel('Bathing Soap')

ax2.plot(x, fwash, 'y.-')
ax2.set_xlabel('Months')
ax2.set_ylabel('Face Wash')

plt.show()
```

```python
#Exercise Question 10: Read all product sales data and show it using the stack plot
ps=df.iloc[:,0:7]
ps.plot(x='month_number', kind='area', stacked=True,
        title='Product Sales per Month')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.show()
```

```python
labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
months=df.month_number.tolist()
ps1=df.facecream.tolist()
ps2=df.facewash.tolist()
ps3=df.toothpaste.tolist()
ps4=df.bathingsoap.tolist()
ps5=df.shampoo.tolist()
ps6=df.moisturizer.tolist()
fig, ax = plt.subplots()
ax.stackplot(months,ps1,ps2,ps3,ps4,ps5,ps6,labels=labels)
ax.legend(loc='upper left')
plt.show()
```

```python
plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(months,ps1,ps2,ps3,ps4,ps5,ps6,
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
