## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
colors=sns.color_palette('twilight_r',25)
```

### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here


QUERY = """
SELECT ConvertedComp
FROM master

"""
hist=pd.read_sql_query(QUERY,conn)

plt.hist(hist,color='r')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master

"""
age=pd.read_sql_query(QUERY,conn)

sns.boxplot(age,color='m')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs
FROM master

"""
scatter=pd.read_sql_query(QUERY,conn)

sns.scatterplot(scatter,y='Age',x='WorkWeekHrs',color='c',marker='+')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master

"""
bubble=pd.read_sql_query(QUERY,conn)
bubble.Age.isnull().sum()
```

```python
bubble.dropna(subset='Age',inplace=True)
px.scatter(bubble,y='WorkWeekHrs',x='CodeRevHrs',size='Age',color='WorkWeekHrs')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear, count(*)
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count(DatabaseDesireNextYear) desc limit 5
"""

df = pd.read_sql_query(QUERY,conn)
df.set_index('DatabaseDesireNextYear',inplace=True)

plt.pie(df['count(*)'],autopct='%1.1f%%',shadow=True,labels=df.index,colors=colors)
plt.show()
```

```python
QUERY = """
SELECT LanguageDesireNextYear, count(*)
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by count(LanguageDesireNextYear) desc limit 5
"""

df = pd.read_sql_query(QUERY,conn)
df.set_index('LanguageDesireNextYear',inplace=True)

plt.pie(df['count(*)'],autopct='%1.1f%%',shadow=True,labels=df.index,colors=colors)
plt.show()
```

```python
QUERY = """
SELECT LanguageWorkedWith, count(*)
FROM LanguageWorkedWith
group by LanguageWorkedWith
order by count(LanguageWorkedWith) desc limit 5
"""

df = pd.read_sql_query(QUERY,conn)
df
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs,Age
FROM master
where Age between 30 and 35

"""

stacked = pd.read_sql_query(QUERY,conn)
stacked=stacked.groupby(['Age'])[['WorkWeekHrs','CodeRevHrs']].median()
stacked.reset_index(inplace=True)
stacked.plot(x='Age',kind='area', stacked=True,
     title='Median Work and Code Review Hours Ages 30 to 35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp,Age
FROM master
where Age between 45 and 60

"""

line = pd.read_sql_query(QUERY,conn)
sns.lineplot(line,x='Age',y='ConvertedComp',color='m')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch,count(*)
FROM master
group by MainBranch
order by count(MainBranch)
"""

barh = pd.read_sql_query(QUERY,conn)
y=barh.iloc[:,0]
w=barh.iloc[:,1]
plt.barh(y,w,color=['r','b'])
plt.show()
```

```python
QUERY = """
SELECT DevType,count(*)
FROM DevType
group by DevType
order by count(DevType)
"""

barh = pd.read_sql_query(QUERY,conn)

y=barh.iloc[:,0]
w=barh.iloc[:,1]
plt.barh(y,w,color=colors)
plt.show()
```
